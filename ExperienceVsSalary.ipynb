```python
# Importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Dataset and Extracting Dependent and Independent Variables
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
y
```

```python
# Visualising the Dataset
dataset.head(10)
```

```python
sns.heatmap(dataset.corr())
```

```python
sns.distplot(dataset['YearsExperience'],bins=10,kde=False)
```

```python
plt.figure(figsize=(10,12))
sns.countplot(x="YearsExperience",data=dataset)
```

```python
plt.figure(figsize=(13,7))
sns.barplot(x='YearsExperience',y='Salary',data=dataset)
```

```python
# Splitting the Dataset into Training set and Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
# Training the data
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train,y_train)
```

```python
# Predicting the Test set results
y_pred = clf.predict(X_test)
y_pred
```

```python
# Visualising the Training set results
plt.scatter(X_train,y_train,color="Blue")
plt.plot(X_train,clf.predict(X_train),color="Red")
plt.title("Salary Vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test,y_test,color="Blue")
plt.plot(X_train,clf.predict(X_train),color="Red")
plt.title("Salary Vs Experience(Testing Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Finding the Residuals
from sklearn import metrics
print('MAE',metrics.mean_absolute_error(y_test,y_pred))
print('MSE',metrics.mean_squared_error(y_test,y_pred))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
